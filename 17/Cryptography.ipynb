```python
# practice of securing information by transforming it into a format unreadable to unauthorized parties.
# standard library and third-party packages provide powerful tools to implement encryption, hashing, etc
# without cryptography:
    # integrity is compromised
    # confidentiality is lost
    # authenticity is questionable

# hash is a fingerprint 
# hashlib module    
    # a hash function is a mathematical function that takes an input and produces a fixed-size output, called a hash value, or hash.
    # changing part of the file/object changes hash.
    # hash algorithms should be:
        # deterministic; same message should always produce the same hash
        # irreversible; should not be feasible to determine original
        # collision resistant
        # fast to compute
```

```python
import hashlib

msg = "Hello, World!"
b_msg = bytes(msg, "utf-8")
print(msg, b_msg)
hash_value = hashlib.sha256(b_msg).hexdigest()
print(hash_value)

msg = "Hello, World?"
b_msg = bytes(msg, "utf-8")
print(msg, b_msg)
hash_value = hashlib.sha256(b_msg).hexdigest()
print(hash_value)
```

```python
# create a file
with open("test.txt", "w") as file:
    file.write("Secret data inside a file! My password is 123456!! Nobody should see it!")

# Open the file for reading
with open("test.txt", "rb") as file:
    file_content = file.read()

# create a hash of file content

hash_value = hashlib.sha256(file_content)
print(hash_value.hexdigest())    

# Open the file for reading
with open("test.txt", "a") as file:
    file.write("New info")


# Open the file for reading
with open("test.txt", "rb") as file:
    file_content = file.read()
    
hash_value = hashlib.sha256(file_content)
print(hash_value.hexdigest()) 
```
